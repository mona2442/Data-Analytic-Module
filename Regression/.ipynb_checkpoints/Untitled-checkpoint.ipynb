```python
import pandas as pd

df = pd.read_csv("Data on permeability.csv")

print(df.shape) #This csv file has 79 rows and 13 columns

df.head(10)  #it prints the first 10 rows of the csv file!
```

```python
%matplotlib inline 
import numpy as np
import pylab as plt
plt.style.use('../settings/plot_style.mplstyle')
```

```python
x_1 = df['density (g/cc)'].values
x_2 = df['Melting'].values
x_3 = df['FFV'].values
x_4 = df['cohesive energy density (Cal/cc)'].values
y = df['alpha_O/N'].values

import numpy as np

def calc_stats(x,y):    
    y_bar = np.mean(y)
    y_std = np.std(x)
    m, b = np.polyfit(x,y,deg=1)
    SST = sum((y - y_bar)**2)
    SSE = sum((y - (m*x+b))**2)
    R2 = (SST - SSE)/SST
    RMSE = np.sqrt(1/len(y)*SSE)
    return y_bar, y_std, m, b, R2, RMSE

fig, axes = plt.subplots(1,4,figsize=(15,4))


axes[0].plot(x_1, y, 'o')
axes[1].plot(x_2, y, 'o')
axes[2].plot(x_3, y, 'o')
axes[3].plot(x_4, y, 'o')


for i, ax in enumerate(axes):

    ax.set_ylim([0, 13])    
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Dataset {}'.format(str(i + 1)))

plt.show()

stats1 = calc_stats(x_1,y)
print("Dataset 1: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats1))
stats2 = calc_stats(x_2,y)
print("Dataset 2: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats2))
stats3 = calc_stats(x_3,y)
print("Dataset 3: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats3))
stats4 = calc_stats(x_4,y)
print("Dataset 4: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats4))
avg, std, m, b, r2, RMSE = stats1
```

```python
fig, axes = plt.subplots(1,4,figsize=(15,4))

yhat_1 = 1.59*x_1 + 2.11
axes[0].plot(x_1, y, 'o')
axes[0].plot(x_1, yhat_1, ls = '-')

yhat_2 = 0.01*x_2 + 0.61
axes[1].plot(x_2, y, 'o')
axes[1].plot(x_2, yhat_2, ls = '-')

yhat_3 = -13.37*x_3 + 6.98
axes[2].plot(x_3, y, 'o')
axes[2].plot(x_3, yhat_3, ls = '-')

yhat_4 = 0.02*x_4 + 1.89
axes[3].plot(x_4, y, 'o')
axes[3].plot(x_4, 0.02* x_4 + 1.89, ls = '-')

for i, ax in enumerate(axes):

    ax.set_ylim([0, 13])    
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Dataset_new {}'.format(str(i + 1)))

plt.show()

stats1 = calc_stats(x_1,yhat_1)
print("Dataset 1: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats1))
stats2 = calc_stats(x_2,yhat_2)
print("Dataset 2: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats2))
stats3 = calc_stats(x_3,yhat_3)
print("Dataset 3: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats3))
stats4 = calc_stats(x_4,yhat_4)
print("Dataset 4: mean={:.2f}, stdev={:.2f}, m={:.2f}, b={:.2f}, R2={:.2f}, RMSE={:.2f}".format(*stats4))
avg, std, m, b, r2, RMSE = stats1
```

```python
fig, axes = plt.subplots(1,2, figsize = (11, 5))
yhat_1 = m*x_1 + b

axes[0].plot(x_1, y, 'o')
axes[0].plot(x_1, yhat_1, ls='-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Dataset 1')

axes[1].plot(y, yhat_1, 'o')
axes[1].plot([min(y), max(y)], [min(y), max(y)], ls='--')
axes[1].set_xlabel('actual data')
axes[1].set_ylabel('predicted value')
axes[1].set_title('Parity Plot')


plt.show()
```

```python
import math
from sklearn.metrics import r2_score

## find the boundaries of X and Y values
bounds = (min(x_1.min(), y.min()) - int(0.1 * y.min()), max(x_1.max(), y.max())+ int(0.1 * y.max()))

# Reset the limits
ax = plt.gca()
ax.set_xlim(bounds)
ax.set_ylim(bounds)
# Ensure the aspect ratio is square
ax.set_aspect("equal", adjustable="box")

plt.plot(x_1,y,"o", alpha=0.5 ,ms=10, markeredgewidth=0.0)

ax.plot([0, 1], [0, 1], "r-",lw=2 ,transform=ax.transAxes)

# Calculate Statistics of the Parity Plot 
mean_abs_err = np.mean(np.abs(x_1-y))
rmse = np.sqrt(np.mean((x_1-y)**2))
rmse_std = rmse / np.std(y)
z = np.polyfit(x_1,y, 1)
y_hat = np.poly1d(z)(x_1)

text = f"$\: \: Mean \: Absolute \: Error \: (MAE) = {mean_abs_err:0.3f}$ \n $ Root \: Mean \: Square \: Error \: (RMSE) = {rmse:0.3f}$ \n $ RMSE \: / \: Std(y) = {rmse_std :0.3f}$ \n $R^2 = {r2_score(y,y_hat):0.3f}$"

plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=10, verticalalignment='top')

# Title and labels 
plt.title("Parity Plot")
plt.xlabel('Ground Truth')
plt.ylabel('Prediction')
```
